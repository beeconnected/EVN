```python
data['Dauer'] = pd.to_numeric(data['Dauer'], errors='coerce')
data = data.dropna()
data = data.drop(data[data['Dauer']== 0].index).reset_index(drop=True)

for y in range(len(data)) :
    if data['Dauer'].values[y] % 60 == 0 :
        data['Dauer'].values[y] = data['Dauer'].values[y] / 60
    else :
        data['Dauer'].values[y] = data['Dauer'].values[y]/60+1
```

```python
#change the tarif to the becotarif
tarifswitch = [('4,9','0.145'),('12,35','0.389'),('1,24','0.09'),('6,4','0.12'),('18,8','0.65'),('31,2','0.71'),('64,35','0.99'),('76,7','1.05')]
dropindex = []
notvalidnumbers = []
altetarife = ['4,9','12,35','1,24','6,4','18,8','31,2','64,35','76,7']

for x in range(len(data['Preis/Einheit'])) :
    try :
        if not altetarife.__contains__(str(data['Preis/Einheit'].values[x])) or not phonenumbers.is_valid_number(phonenumbers.parse(data['Absender'].values[x], "DE")):
            dropindex.append(x)
        else :
            for k, v in tarifswitch:
                data['Preis/Einheit'].values[x] =str(data['Preis/Einheit'].values[x]).replace(k,v)

    except NumberParseException:
        print("** this is not a valid number ** --> "+ data['Absender'].values[x])
        dropindex.append(x)
        notvalidnumbers.append(data['Absender'].values[x])

data = data.drop(dropindex).reset_index(drop=True)
data['Preis/Einheit'] = pd.to_numeric(data['Preis/Einheit'], errors="coerce")
data = data.dropna()
```
